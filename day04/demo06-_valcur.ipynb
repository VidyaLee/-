```python
#验证曲线

#小汽车分类   随机森林分类器
import numpy as np
import sklearn.preprocessing as sp
import sklearn.ensemble as se   #集成
import sklearn.model_selection as ms


#读取数据样本，对每一列进行标签编码
#整理样本空间，输入集与输出集
data=[]
with open('../ml_data/car.txt','r') as f:
    for line in f.readlines():
        data.append(line.split(','))
data=np.array(data)

train_x,train_y=[],[]
encoders=[]  #存储所有的标签编码器，以后使用

for index ,row in enumerate(data.T):
    encoder=sp.LabelEncoder()   #encoder 可以将数据集中的文本转化成0或1的数值。
    if index<(len(data.T)-1):    #添加到输入
        train_x.append(encoder.fit_transform(row))
    else:    #添加到输出
        train_y=encoder.fit_transform(row)
train_x=np.array(train_x).T
train_x=np.array(train_x)
```

```python
#训练随机森林分类器模型
model=se.RandomForestClassifier(max_depth=6,n_estimators=200,random_state=7)

#验证曲线选择最优的超参数n_estimators
train_scores,test_score=ms.validation_curve(model,train_x,train_y,'n_estimators',np.arange(50,550,50),cv=5)
print(train_scores)
print('*'*100)
print(train_scores.mean(axis=1))
```

```python
#绘制验证曲线结果
import matplotlib.pyplot as mp
mp.grid(linestyle=':')
mp.plot(np.arange(50,550,50),test_score.mean(axis=1),'o-',color='dodgerblue',label='validation curve')

mp.legend()
mp.show()
```

```python
model=se.RandomForestClassifier(max_depth=6,n_estimators=200,random_state=7)
#验证曲线选择最优的超参数n_estimators
train_scores,test_score=ms.validation_curve(model,train_x,train_y,'n_estimators',np.arange(10,200,10),cv=5)
#绘制验证曲线结果
import matplotlib.pyplot as mp
mp.grid(linestyle=':')
mp.plot(np.arange(10,200,10),test_score.mean(axis=1),'o-',color='dodgerblue',label='validation curve')

mp.legend()
mp.show()
```
